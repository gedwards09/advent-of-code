```python
class IGraphNode:
    
    def getData(self):
        return self._data
    
    def setData(self, data):
        self._data = data
        
    def addChild(self, child):
        self._children.append(child)
        
    def iterateChildren(self):
        for child in self._children:
            yield child
```

```python
class DirectedWeightedGraphNode(IGraphNode):
    
    def __init__(self, data):
        self._data = data
        self._children = {}
        
    #Override
    def addChild(self, child, weight = 1):
        self._children[child] = weight
        
    def iterateChildWeights(self):
        for child in self._children:
            yield child, self._children[child]
    
    def print_(self):
        print(str(self._data) + ': ' \
              + ', '.join([':'.join([str(child.getData()), str(self._children[child])]) \
                                     for child in self._children]))
```

```python
class DirectedWeightedGraph:
    
    def __init__(self):
        self._nodeIndex = {}
        self._edges = {}
        
    def addNode(self, name):
        node = DirectedWeightedGraphNode(name)
        self._nodeIndex[name] = node
        
    def getNode(self, name):
        if name not in self._nodeIndex:
            raise Exception("Node {0} not found!".format(name))
        return self._nodeIndex[name]
    
    def addEdge(self, sourceName, destinationName, directionType, weight = 1):
        source = self._nodeIndex[sourceName]
        destination = self._nodeIndex[destinationName]
        if directionType == 1 or directionType == 0:
            source.addChild(destination, weight)
        if directionType == -1 or directionType == 0:
            destination.addChild(source, weight)
    
    def iterateNodes(self):
        for key in self._nodeIndex:
            yield self._nodeIndex[key]
            
    def isNode(self, name):
        return name in self._nodeIndex
    
    def print_(self):
        for name in self._nodeIndex:
            self._nodeIndex[name].print_()
```

```python
class ForestMap:
    
    s_startCoord = (0, 1)
    
    def __init__(self, text):
        self._map = text.split('\n')
        self._yBound = len(self._map)
        self._xBound = len(self._map[0])
        self._graph = self.initGraph()
        
    def getGraph(self):
        return self._graph
        
    def getChar(self, coord):
        i, j = coord
        return self._map[i][j]
        
    def getStartCoord(self):
        return ForestMap.s_startCoord
    
    def getEndCoord(self):
        return (self._yBound - 1, self._xBound - 2)
        
    def initGraph(self):
        graph = DirectedWeightedGraph()
        startCoord = self.getStartCoord()
        endCoord = self.getEndCoord()
        graph.addNode(startCoord)
        graph.addNode(endCoord)
        junctionPathsFollowed = {startCoord:[], endCoord:[]}
        for i in range(self._yBound):
            for j in range(self._xBound):
                if self.isOnPath(i, j) and self.isJunction(i, j):
                    junctionPathsFollowed[(i,j)] = []
                    graph.addNode((i, j))
        graph = self.addGraphEdges(graph, junctionPathsFollowed)
        return graph
    
    def isOnPath(self, i, j):
        char = self._map[i][j]
        return char == '.' or char == 'v' or char == '>'
    
    def isJunction(self, i, j):
        return self.getDegree(i, j) >= 3
    
    def getDegree(self, i, j):
        degree = 0
        for k, l in self.iterateAdjacentPathCoordinates(i, j):
            if self.isOnPath(k, l):
                degree += 1
        return degree
    
    def iterateAdjacentPathCoordinates(self, y, x):
        for d in [-1, 1]:
            if self.isValid(y + d, x) and self.isOnPath(y + d, x):
                yield y + d, x
            if self.isValid(y, x + d) and self.isOnPath(y, x + d):
                yield y, x + d
    
    def isValid(self, i, j):
        return 0 <= i and i < self._yBound and 0 <= j and j < self._xBound
    
    def addGraphEdges(self, graph, junctionPathsFollowed):
        junctions = junctionPathsFollowed.copy()
        for startCoord in junctions:
            i, j = startCoord
            node = graph.getNode(startCoord)
            for coord in self.iterateAdjacentPathCoordinates(i, j):
                if coord in junctionPathsFollowed[startCoord]:
                    continue
                # consider whether to handle nodes connected by multiple edges
                endCoord, steps, directionType = self.followPathSegment(startCoord, coord, \
                                                                        junctionPathsFollowed)
                if endCoord == None:
                    continue
                graph.addEdge(startCoord, endCoord, directionType, steps)
        return graph
            
    # returns: endCoord: the coordinate of the ending junction
    #          steps: the number of steps to the ending junction
    #          directionType: the direction type of path segment:
    #                         0 both directions
    #                         1 forwards only
    #                        -1 backwards only
    def followPathSegment(self, junctionCoord, pathStartCoord, junctions):
        directionType = 0
        steps = 1
        prevCoord = junctionCoord
        currentCoord = pathStartCoord
        isDeadEnd = False
        while not currentCoord in junctions and not isDeadEnd:
            isDeadEnd = True
            i, j = currentCoord
            for nextCoord in self.iterateAdjacentPathCoordinates(i ,j):
                if nextCoord == prevCoord:
                    continue
                isDeadEnd = False
                curChar = self.getChar(nextCoord)
                if self.isSlopedPath(curChar):
                    if currentCoord == self.getDestinationCoordinate(nextCoord):
                        directionType = ForestMap.setDirectionType(directionType, -1)
                    else:
                        directionType = ForestMap.setDirectionType(directionType, 1)
                steps += 1
                prevCoord = currentCoord
                currentCoord = nextCoord
                break
        if currentCoord in junctions:
            junctions[junctionCoord].append(pathStartCoord)
            junctions[currentCoord].append(prevCoord)
            return currentCoord, steps, directionType
        # dead end
        return None, None, None
    
    def isSlopedPath(self, char):
        return char != '.'
                    
    def getDestinationCoordinate(self, coord):
        char = self.getChar(coord)
        if char != '>' and char != 'v':
            return None
        i, j = coord
        if char == 'v':
            y, x = i + 1, j
        elif char == '>':
            y, x = i, j + 1
        if self.isOnPath(y, x):
            return (y, x)
        else:
            return None
    
    def setDirectionType(oldType, newType):
        if oldType == 0 or oldType == newType:
            return newType
        raise Exception("Encountered path with conflicting direction types")
        
    def print_(self):
        self._graph.print_()
```

```python
class PathSearch:
    
    def __init__(self, text):
        self._forestMap = self.initForestMap(text)
        self._startNode = self.initStartNode()
        self._endNode = self.initEndNode()
        self._memo = {}
        
    def initForestMap(self, text):
        return ForestMap(text)
        
    def initStartNode(self):
        return self._forestMap.getGraph().getNode(self._forestMap.getStartCoord())
    
    def initEndNode(self):
        return self._forestMap.getGraph().getNode(self._forestMap.getEndCoord())
        
    def getMaximumPathLength(self):
        startCoord = self._forestMap.getStartCoord()
        root = self._forestMap.getGraph().getNode(startCoord)
        return self.longestPathSearch(root)
    
    def longestPathSearch(self, node):
        return self.rLongestPathSearch(node)
    
    def rLongestPathSearch(self, node, ancestors = {}):
        ancestors[node.getData()] = ''
        key = (self.makeKey(ancestors), node.getData())
        if key in self._memo:
            return self._memo[key]
        if node == self._endNode:
            self._memo[key] = 0
            return 0
        curMax = 0
        for child, weight in node.iterateChildWeights():
            if child.getData() in ancestors:
                continue
            out = self.rLongestPathSearch(child, ancestors.copy())
            if out == None:
                continue
            out += weight
            if out > curMax:
                curMax = out
        # no children remaining, didn't reach end node
        if curMax == 0:
            self._memo[key] = None
            return None
        self._memo[key] = curMax
        return curMax
        
    def makeKey(self, dic):
        cur = 0
        for key in dic:
            cur ^= hash(str(key))
        return cur
```

```python
pathSearch = PathSearch(text)
print(pathSearch.getMaximumPathLength())
```

```python
def addNodes(G, forestMap):
    G.add_nodes_from([node.getData() for node in forestMap.getGraph().iterateNodes()])
    return G

def addEdges(G, forestMap):
    for node in forestMap.getGraph().iterateNodes():
        for child, weight in node.iterateChildWeights():
            G.add_edge(node.getData(), child.getData(), label = weight)
    return G

import networkx as nx
import matplotlib.pyplot as plt
forestMap = ForestMap(text)
G = nx.Graph()
G = addNodes(G, forestMap)
G = addEdges(G, forestMap)
edge_labels = nx.get_edge_attributes(G, "label")
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos=pos)
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels = edge_labels)
plt.show()
```
