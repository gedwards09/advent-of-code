```python
dic = {}
def init_dic(dic):
    dic["red"] = 0
    dic["green"] = 0
    dic["blue"] = 0
    return dic

def get_power(dic):
    return dic["red"] * dic["green"] * dic["blue"]

arr = text.split('\n')
s = 0
for line in arr:
    dic = init_dic(dic)
    post = line.split(':')
    gamenum = int(post[0][5:])
    for set in post[1].split(';'):
        for dice in set.split(','):
            el = dice.split(' ')
            count = int(el[1])
            type = el[2]
            if count > dic[type]:
                dic[type] = count
    s += get_power(dic)
print(s)
```
