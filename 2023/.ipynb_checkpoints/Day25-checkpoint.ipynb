```python
import networkx as nx

def getNodesAndEdges(text):
    index = {}
    nodes = []
    edges = []
    for line in text.split('\n'):
        pieces = line.split(': ')
        addNodeToList(nodes, index, pieces[0])
        for name in pieces[1].split(' '):
            addNodeToList(nodes, index, name)
            edges.append((pieces[0], name))
    return nodes, edges

def addNodeToList(arr, index, nodeName):
    if not nodeName in index:
        arr.append(nodeName)
        index[nodeName] = ''
            
g = nx.Graph()
nodes, edges = getNodesAndEdges(text)
g.add_nodes_from(nodes)
g.add_edges_from(edges)
nx.draw_kamada_kawai(g, with_labels = True)
```

```python
class Node:
    
    def __init__(self, data):
        self._data = data
        self._next = {}
        
    def getData(self):
        return self._data
    
    def iterateNext(self):
        for node in self._next:
            yield node
    
    def addEdge(self, node):
        if not node in self._next:
            self._next[node] = ''
            node._next[self] = ''
            
    def isAdjacent(self, node):
        return node in self._next
```

```python
class Path:
    
    def __init__(self, end, nodes = {}):
        self._nodes = nodes
        self._end = end
        
    def getEnd(self):
        return self._end
    
    def step(self, node):
        self._nodes[self._end] = ''
        self._end = node
        
    def iterate(self):
        for node in self._nodes:
            yield node
    
    def copy(self):
        return Path(self._end, self._nodes.copy())
```

```python
class Apparatus:
    
    s_connectivityDegree = 4
    
    def __init__(self, text):
        self._nodeIndex = {}
        self._size = 0
        self.initNodes(text)
        
    def initNodes(self, text):
        for line in text.split('\n'):
            head, tail = line.split(': ')
            headNode = self.makeNode(head)
            for name in tail.split(' '):
                node = self.makeNode(name)
                headNode.addEdge(node)
                    
    def makeNode(self, data):
        if not data in self._nodeIndex:
            self._nodeIndex[data] = Node(data)
            self._size += 1
        return self._nodeIndex[data]
    
    def getProductOfComponentSize(self):
        count = self.partitionGraphComponentSize()
        return count * (self._size - count)
                    
    def partitionGraphComponentSize(self):
        start = list(self._nodeIndex.values())[0]
        visited = {start:''}
        queue = [start]
        count = 1
        while len(queue) > 0:
            current = queue.pop(0)
            for node in current.iterateNext():
                if not node in visited and self.isSameComponent(current, node):
                    visited[node] = ''
                    queue.append(node)
                    count += 1
        return count
            
    def isSameComponent(self, start, node):
        return self.hasNDistinctPaths(start, node, Apparatus.s_connectivityDegree)
        
    def hasNDistinctPaths(self, start, end, n):
        if n <= 0:
            return True
        if start.isAdjacent(end):
                return self.rHasNPaths(start, end, n - 1, excluded = {end:''})
        return self.rHasNPaths(start, end, n)
    
    def rHasNPaths(self, start, end, n, excluded = None):
        if n <= 0:
            return True
        if excluded == None:
            excluded = {}
        visited = {node:'' for node in start.iterateNext() if node != end and not node in excluded}
        visited[start] = ''
        queue = [Path(node, {start:''}) \
                 for node in start.iterateNext() \
                 if node != end and not node in excluded]
        path = None
        done = False
        while not done and len(queue) > 0:
            path = queue.pop(0)
            current = path.getEnd()
            for node in current.iterateNext():
                if node == end:
                    done = True
                    path.step(node)
                    break
                elif not node in visited and not node in excluded:
                    visited[node] = ''
                    copyPath = path.copy()
                    copyPath.step(node)
                    queue.append(copyPath)
        if done:
            for node in path.iterate():
                excluded[node] = ''
            return self.rHasNPaths(start, end, n - 1, excluded = excluded)
        return False
```
