```python
class GardenPlot:
    
    def __init__(self, text):
        self._map = text.split('\n')
        self._yBound = len(self._map)
        self._xBound = len(self._map[0])
        self._start = self.initStart()
        
    def initStart(self):
        for i in range(self._yBound):
            for j in range(self._xBound):
                if self._map[i][j] == 'S':
                    return (i, j)
                
    def getMap(self):
        return self._map
    
    def getYBound(self):
        return self._yBound
    
    def getXBound(self):
        return self._xBound
    
    def getStart(self):
        return self._start
    
    def count(self, steps = 64):
        dic = self.generatePlotPositions(steps)
        ct = 0
        for key in dic:
            ct += 1
        return ct
        
    def generatePlotPositions(self, steps):
        if steps == 0:
            return {self._start:''}
        dic = {}
        for coord in self.generatePlotPositions(steps - 1):
            for y, x in GardenPlot.enumeratePositions(coord):
                if self.isValid((y, x)):
                    plot = self._map[y][x]
                    if plot == '.' or plot == 'S':
                        dic[(y, x)] = ''
        return dic
    
    def enumeratePositions(coord):
        y, x = coord
        for dy, dx in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            yield y + dy, x + dx
        
    def isValid(self, coord):
        y, x = coord
        if 0 <= y and y < self._yBound and 0 <= x and x < self._xBound:
            return True
```

```python
text = """...........
.....###.#.
.###.##..#.
..#.#...#..
....#.#....
.##..S####.
.##..#...#.
.......##..
.##.#.####.
.##..##.##.
..........."""
```

```python
gardenPlot = GardenPlot(text)
gardenPlot.count(6)
```

```python
class InfiniteGardenPlot(GardenPlot):
    
    #Override
    def count(self, steps = 26501365):
        count, _ = self.generatePlotPositions(steps)
        return count
        
    #Override
    def generatePlotPositions(self, steps):
        map_ = self.getMap()
        yBound = self.getYBound()
        xBound = self.getXBound()
        previousPositions = {}
        previousCount = 0
        positions = {self.getStart():''}
        count = 1
        for i in range(steps):
            nextPositions = {}
            nextCount = previousCount
            for coord in positions:
                for y, x in GardenPlot.enumeratePositions(coord):
                    if not (y, x) in previousPositions:
                        i = InfiniteGardenPlot.getInBoundsCoordinates(y, yBound)
                        j = InfiniteGardenPlot.getInBoundsCoordinates(x, xBound)
                        plot = map_[i][j]
                        if (plot == '.' or plot == 'S') and not (y, x) in nextPositions:
                            nextPositions[(y, x)] = ''
                            nextCount += 1
            previousPositions = positions
            previousCount = count
            positions = nextPositions
            count = nextCount
        return count, positions
                
    def getInBoundsCoordinates(y, yBound):
        y %= yBound
        if y < 0:
            y += yBound
        return y
        
```

```python
infiniteGardenPlot = InfiniteGardenPlot(text)
for i in [0, 1, 2, 3, 6, 10, 50, 100, 500, 1000, 5000]:
    print(i, infiniteGardenPlot.count(i))
```
